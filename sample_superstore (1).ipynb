```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading dataset
```

```python
df = pd.read_csv("Sample - Superstore.csv")
```

```python
#renaming some columns
```

```python
df.columns=df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace('ID', 'Id')
```

```python
df.columns
```

```python
#checking missing values
```

```python
df.isnull().sum()
```

```python
#checking duplicates
```

```python
df.duplicated()
```

```python
#checking data types
```

```python
df.dtypes
```

```python
#changing some datatypes
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
```

```python
#analyzing some columns
```

```python
df['Ship_Mode'].value_counts()
```

```python
df['Segment'].value_counts()
```

```python
df['Country'].value_counts()
```

```python
df['City'].value_counts()
```

```python
df['State'].value_counts()
```

```python
df['Region'].value_counts()
```

```python
df.dtypes
```

```python
#changing id columns to category 
```

```python
df['Order_Id'] = df['Order_Id'].astype('category')
df['Customer_Id'] = df['Customer_Id'].astype('category')
df['Product_Id'] = df['Product_Id'].astype('category')
```

```python
#changing columns to categoryy together
```

```python
cat_cols = ['Ship_Mode', 'Segment', 'Country', 'City', 
            'State', 'Region', 'Category', 'Sub-Category']

for col in cat_cols:
    df[col] = df[col].astype('category')
```

```python
#changing postal code to object
```

```python
df['Postal_Code'] = df['Postal_Code'].astype('str')
```

```python
df.dtypes
```

```python
#making new columns for analysis
```

```python
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Month_Name'] = df['Order_Date'].dt.month_name()
```

```python
#checking some values
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_order_value = df['Sales'].mean()
total_orders = df['Order_Id'].nunique()
```

```python
total_sales, total_profit, avg_order_value, total_orders
```

```python
#grouping column region by sales
```

```python
top_region = df.groupby('Region',observed='True')['Sales'].sum().sort_values(ascending=False)
```

# Data Analysis

```python
#grouping sales by category
```

```python
sales_category = df.groupby('Category',observed=True)['Sales'].sum().sort_values(ascending=False)
```

```python
#grouping sales by sub-category and knowing first 10 values
```

```python
top_subcat = df.groupby('Sub-Category',observed=True)['Sales'].sum().sort_values(ascending=False)
top_subcat.head(10)
```

```python
#grouping sales by region
```

```python
sales_region = df.groupby('Region',observed=True)['Sales'].sum()
sales_region
```

```python
#grouping profit by category
```

```python
profit_category = df.groupby('Category',observed=True)['Profit'].sum()
profit_category
```

```python
#checking loss making products and seeing first 5 products making loss
```

```python
loss_products = df[df['Profit'] < 0]
loss_products.head()
```

```python
#grouping sales by month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

# Data Visualization

```python
#plotting sales trend chart
```

```python
plt.figure(figsize=(10,5))

plt.plot(monthly_sales.index, monthly_sales.values)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.tight_layout()
```

```python
plt.savefig("sales_trend.png", bbox_inches='tight')
```

```python
#plotting sales by category
```

```python
plt.figure(figsize=(8,5))

plt.bar(sales_category.index, sales_category.values)
plt.title('Sales by Category')
plt.xlabel("Category")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
plt.savefig("sales_by_category.png", bbox_inches='tight')
```

```python
#plotting top 10 subcategory
```

```python
plt.figure(figsize=(12,6))

top10 = top_subcat.head(10)
plt.bar(top10.index, top10.values)
plt.title("Top 10 Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.savefig("top_10_subcategory.png", bbox_inches='tight')
```

```python
#plotting sales by region
```

```python
plt.figure(figsize=(8,8))

plt.pie(
    sales_region.values,
    labels=sales_region.index,
    autopct='%1.1f%%'
)
plt.title("Sales Distribution by Region")
plt.tight_layout()
plt.savefig("sales_by_region.png", bbox_inches='tight')  # save BEFORE show
plt.show()
plt.close()
```

```python
#plotting profit by category
```

```python
plt.figure(figsize=(10,5))

plt.bar(profit_category.index, profit_category.values)
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.tight_layout()
plt.savefig("profit_by_category.png", bbox_inches='tight')  # save BEFORE show
plt.show()
plt.close()
```

```python
#plotting sales vs profit
```

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.tight_layout()
plt.savefig("sales_vs_profit.png", bbox_inches='tight')  # save BEFORE show
plt.show()
plt.close()

```
